```python
# importing all necessary libraires
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import os
```

```python
# DataSet
# Dataset
data = {
    'x1': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y1': [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68],
    'x2': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y2': [9.14, 8.14, 8.74, 8.77, 9.26, 8.1, 6.13, 3.1, 9.13, 7.26, 4.74],
    'x3': [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5],
    'y3': [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73],
    'x4': [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8],
    'y4': [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.5, 5.56, 7.91, 6.89]
}
```

```python
# Task 1: Constructing data frames
df1 = pd.DataFrame(data, columns=['x1', 'y1'])
df2 = pd.DataFrame(data, columns=['x2', 'y2'])
df3 = pd.DataFrame(data, columns=['x3', 'y3'])
df4 = pd.DataFrame(data, columns=['x4', 'y4'])
```

```python
# Task 2: Descriptive statistics
print("Descriptive Statistics for Dataset 1:")
print(df1.describe())
print("\nDescriptive Statistics for Dataset 2:")
print(df2.describe())
print("\nDescriptive Statistics for Dataset 3:")
print(df3.describe())
print("\nDescriptive Statistics for Dataset 4:")
print(df4.describe())
```

```python
# Task 3: Scatterplots
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.scatter(df1['x1'], df1['y1'])
plt.title('Dataset I')

plt.subplot(2, 2, 2)
plt.scatter(df2['x2'], df2['y2'])
plt.title('Dataset II')

plt.subplot(2, 2, 3)
plt.scatter(df3['x3'], df3['y3'])
plt.title('Dataset III')

plt.subplot(2, 2, 4)
plt.scatter(df4['x4'], df4['y4'])
plt.title('Dataset IV')

plt.tight_layout()
plt.show()
```

```python
# Task 4: Explanation
'''
The Anscombe's quartet demonstrates that descriptive statistics alone may not be sufficient to capture the differences
between datasets. Although the descriptive statistics (mean, variance, etc.) are similar for all four datasets,
the scatterplots reveal distinct patterns, highlighting the importance of visualizing data to understand its
underlying structure and relationships.
'''
```

## Problem # 2


```python
# Loading the dataset
data_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv"
df = pd.read_csv(data_url)
```

```python
# Task 1: Scatter plot of median earnings vs. unemployment rate
plt.scatter(df["Median"], df["Unemployment_rate"])
plt.xlabel('Median Earnings')
plt.ylabel('Unemployment Rate')
plt.title('Median Earnings vs. Unemployment Rate')
plt.show()

'''
There is a weak correlation between earnings and unemployment. While petroleum engineering majors
have high earnings and low unemployment, other majors with high earnings may not necessarily have
low unemployment rates.
'''
```

```python
# Task 2: Bar plot for majors with median salary above $60,000
high_earning_majors = df[df["Median"] > 60000]
high_earning_majors_subset = high_earning_majors[["Major", "P25th", "Median", "P75th"]]
high_earning_majors_subset.set_index("Major").plot(kind='bar', figsize=(12, 8))
plt.title('Earnings for Majors with Median Salary > $60,000')
plt.ylabel('Earnings')
plt.show()

'''
Yes, petroleum engineering majors have the highest median earnings among recent graduates, but there
are other majors with high earnings as well, such as actuarial science, court reporting, and public policy.
'''
```

## Problem 3

```python
# Task 1. Reloading user_reviews.csv into a dataframe named df
df = pd.read_csv('user_reviews.csv')

# Task 2. Adding the column "len_text"
df['len_text'] = df['text'].apply(len)

# Task 3a. Creating the "super category" column using for loop
for i in range(len(df)):
    if df.loc[i, 'len_text'] > 1000 or df.loc[i, 'grade'] >= 9:
        df.loc[i, 'super category'] = 'expert reviewer'
    elif df.loc[i, 'len_text'] > 1000 and df.loc[i, 'grade'] <= 1:
        df.loc[i, 'super category'] = 'opposed reviewer'
    else:
        df.loc[i, 'super category'] = 'neutral reviewer'

# Saving the dataframe to a file with a unique name
df.to_csv('user_reviews_category_for_loop.csv', index=False)

# Task 3b. Creating the "super category" column using while loop
df = pd.read_csv('user_reviews.csv')  # Reload the original data
df['len_text'] = df['text'].apply(len)
i = 0
while i < len(df):
    if df.loc[i, 'len_text'] > 1000 or df.loc[i, 'grade'] >= 9:
        df.loc[i, 'super category'] = 'expert reviewer'
    elif df.loc[i, 'len_text'] > 1000 and df.loc[i, 'grade'] <= 1:
        df.loc[i, 'super category'] = 'opposed reviewer'
    else:
        df.loc[i, 'super category'] = 'neutral reviewer'
    i += 1

# Saving the dataframe to a file with a unique name
df.to_csv('user_reviews_category_while_loop.csv', index=False)

# Task 3c. Creating the "super category" column using vectorization
df = pd.read_csv('user_reviews.csv')  # Reload the original data
df['len_text'] = df['text'].apply(len)
df['super category'] = 'neutral reviewer'
df.loc[(df['len_text'] > 1000) | (df['grade'] >= 9), 'super category'] = 'expert reviewer'
df.loc[(df['len_text'] > 1000) & (df['grade'] <= 1), 'super category'] = 'opposed reviewer'

# Saving the dataframe to a file with a unique name
df.to_csv('user_reviews_category_vectorization.csv', index=False)
```

## Porblem # 4

```python
import sqlite3

# Connect to the SQLite database (creates a new one if it doesn't exist)
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

# Clearing the tables if they exist
cursor.execute('DROP TABLE IF EXISTS locations')
cursor.execute('DROP TABLE IF EXISTS departments')

# Task 1a: Create tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS locations
(
location_id INT(3),
street_address VARCHAR(50) UNIQUE NOT NULL,
postal_code VARCHAR(10) NOT NULL,
city VARCHAR(50) NOT NULL,
state_province VARCHAR(50) NOT NULL,
country VARCHAR(50) NOT NULL,
constraint l_lid_pk PRIMARY KEY (location_id)
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS departments
(
department_id INT(4),
department_name VARCHAR(20) UNIQUE NOT NULL,
location_id INT(3),
constraint d_did_pk PRIMARY KEY (department_id),
constraint d_lid_fk FOREIGN KEY (location_id) REFERENCES locations(location_id)
)
''')

# Task 1b: Insert data
locations_data = [
    (100, '2 Nice Road', 'N2 7TH', 'London', 'Greater London', 'UK'),
    (200, '23 Pretty Road', 'BS1 8FD', 'Bristol', 'Bristol County', 'UK'),
    (300, '26 Great Street', 'BN1 4BF', 'Brighton', 'Sussex', 'UK'),
    (400, '143 Lovely Road', 'CB1 2NV', 'Cambridge', 'Cambridgeshire', 'UK')
]

departments_data = [
    (10, 'IT', 100),
    (20, 'Operations', 200),
    (30, 'Sales', 300),
    (40, 'Marketing', 200),
    (50, 'Management', None)
]

cursor.executemany('INSERT INTO locations VALUES (?, ?, ?, ?, ?, ?)', locations_data)
cursor.executemany('INSERT INTO departments VALUES (?, ?, ?)', departments_data)

# Commit changes and close connection
conn.commit()
conn.close()

# Task 2: Display departments and their locations
conn = sqlite3.connect('company.db')
cursor = conn.cursor()

# Display all departments and their locations
cursor.execute('''
SELECT d.department_name, l.location_id, l.city
FROM departments d
LEFT JOIN locations l ON d.location_id = l.location_id
''')

print("Departments and their Locations:")
for row in cursor.fetchall():
    print(row)

# Display locations with no departments
cursor.execute('''
SELECT l.location_id, l.city
FROM locations l
LEFT JOIN departments d ON l.location_id = d.location_id
WHERE d.department_id IS NULL
''')

print("\nLocations with No Departments:")
for row in cursor.fetchall():
    print(row)

# Close connection
conn.close()
```
